```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(nx.__version__)
```

```python
n = 7
G = nx.empty_graph(n)
```

```python
nx.draw(G)
```

```python
nx.draw(G, with_labels=True)
```

```python
Full_graph = nx.complete_graph(n)
nx.draw(Full_graph, with_labels=True)
```

```python
Full_graph_D = nx.DiGraph(Full_graph)
nx.draw(Full_graph_D, with_labels=True)
```

```python
#Star Graph
St = nx.star_graph(n)
nx.draw(St, with_labels=True)
```

```python
#Random Tree
Rt = nx.random_tree(n*2)
nx.draw(Rt)
```

```python
#Balanced tree
Bt = nx.balanced_tree(2, 3)
nx.draw(Bt)
```

```python
BG = nx.balanced_graph(3, 3)
nx.draw(BG)
```
